```python
import numpy as np
import pandas as pd
df_churn=pd.read_csv("/content/drive/MyDrive/Churn_Modelling.csv")
df_churn
```

```python
df_churn.isnull().sum()
```

```python
df_churn.drop("RowNumber",axis=1,inplace=True)
```

```python
df_churn.drop("CustomerId",axis=1,inplace=True)
df_churn
```

```python
d={"Male":0, "Female":1}
df_churn["Gender"]=df_churn["Gender"].map(d)
```

```python
from sklearn.preprocessing import LabelEncoder
for col in df_churn.columns:
  le=LabelEncoder()
  df_churn[col]=le.fit_transform(df_churn[col])
df_churn
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.heatmap(df_churn.corr(), annot=True)
plt.show()
```

```python
features=["Age","Gender","Balance"]
x=df_churn[features]
y=df_churn["Exited"]
```

```python
from sklearn.feature_selection import chi2
chi_scores=chi2(x,y)
chi_scores
```

```python
chi_values=pd.Series(chi_scores[0],index=x.columns)
chi_values.sort_values(ascending=False,inplace=True)
chi_values.plot.bar()
```

```python
p_values=pd.Series(chi_scores[1],index=x.columns)
p_values.sort_values(ascending=False,inplace=True)
p_values.plot.bar()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
np.ascontiguousarray(pca.fit_transform(x))
```

```python
pca.explained_variance_ratio_
```
